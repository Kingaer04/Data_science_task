```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/youtube_api.csv')
```

```python
df.set_index(np.arange(1,len(df)+1), inplace=True)
```

```python
df.head(10)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(20,10))

plt.scatter(df['view_count'], df['likes'], s=100, edgecolor='Black', c=df['ratio'], cmap = 'summer', alpha = 0.7)
cbar = plt.colorbar()
cbar.set_label('Highest views')
plt.xscale('log')
plt.yscale('log')
```
